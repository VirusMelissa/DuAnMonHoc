```python
#dùng kiểu nạp chồng, có thể đưa nhiều giá trị nhưng chỉ duy nhất 1 hàm

def add(*args):
    if len(args) == 3:
        return "Chu vi tam giác = {0}".format(sum(args))
    elif len(args) == 4:
        return "Chu vi tứ giác = {0}".format(sum(args))
    
if __name__ == '__main__':
    a=1
    b=2
    c=3
    print(add(a,b))
    print(add(a,b,c))
    print(add(a,b,c,5))
    
```

# Tham số tùy biến dạng **args

```python
#dùng kiểu nạp chồng, có thể đưa nhiều giá trị (đặc biệt phải có keyword cho mỗi giá trị) nhưng chỉ duy nhất 1 hàm

def ham(**kwargs):
    print(type(kwargs))
    print(kwargs)
    
if __name__ == '__main__':
    ham(a=1,b=2,x=3)
```

```python
def ham(**kwargs):
    a = kwargs.get('a')
    b = kwargs.get('b')
    c = kwargs.get('c')
    d = kwargs.get('d')
    size = len(kwargs)
    
    if a and b and c and size == 3:
        print("Tam giác có các cạnh: {0} _ {1} _ {2}".
              format(kwargs['a'], kwargs['b'], kwargs['c']))
        print("Chu vi tam giác là {0}".
             format(kwargs['a'] + kwargs['b'] + kwargs['c']))
    
if __name__ == '__main__':
    ham(a=1,b=2,c=3)
    ham(a=1,b=2,c=3,d=4)
```

# Tính chu vi và diện tích

```python
def tinh(*args, **kwargs):
    hinh = kwargs.get('hinh')
    tinh = kwargs.get('tinh')
    soCanh = len(args)
    
    if hinh =='vuong' and tinh == 'cv' and soCanh == 1:
        return args[0]*4
    elif hinh =='vuong' and tinh == 'dt' and soCanh==1:
        return args[0]*2
    elif hinh =='tron' and tinh == 'cv' and soCanh==1:
        return sum(args)*2
    elif hinh =='chu_nhat' and tinh == 'cv' and soCanh==2:
        return sum(args)*2
    elif hinh =='chu_nhat' and tinh == 'cv_dt' and soCanh==2:
        return sum(args)*2, args[0]*args[1]
    
if __name__ == '__main__':
    print(tinh(10, hinh='vuong',tinh='cv'))
    print(tinh(50, hinh='vuong',tinh='dt'))
    print(tinh(18, hinh='tron',tinh='cv'))
    print(tinh(20, 30, hinh='chu_nhat', tinh='cv'))
    print(tinh(20, 30, hinh='chu_nhat', tinh='cv_dt'))
```

# Bài tập QLNV cơ bản

```python
def taoNVVP(maNV: int, hoTen: str, luongCB: float, soNG: int) -> dict:
    d = {
        'maNV': maNV,
        'hoTen': hoTen,
        'luongCB': luongCB,
        'soNG': soNG,
        'luongHT': 0
    }
    return d

def taoNVBH(maNV: int, hoTen: str, luongCB: float, soSP: int) -> dict:
    d = {
        'maNV': maNV,
        'hoTen': hoTen,
        'luongCB': luongCB,
        'soSP': soSP,
        'luongHT': 0
    }
    return d

# 1. Khởi tạo tự động các nhân viên
def khoiTaoNV(dsNV: list):
    dsNV.append(taoNVVP(123, 'Nguyễn Văn A', 5_600_000, 23))
    dsNV.append(taoNVVP(124, 'Nguyễn Văn B', 6_900_000, 123))
    dsNV.append(taoNVVP(125, 'Nguyễn Văn C', 7_600_000, 50))
    dsNV.append(taoNVBH(126, 'Nguyễn Văn D', 8_900_000, 120))
    dsNV.append(taoNVBH(127, 'Nguyễn Văn E', 10_600_000, 67))
    dsNV.append(taoNVBH(128, 'Nguyễn Văn F', 5_600_000, 210))
    
# 2. Xuất nhân viên
def printNV(dsNV: list): #kiểu mong muốn list
    for nv in dsNV:
        print(list(nv.values()))
        
# 3. Tính lương
def tinhluongHT(ds: list):
    for nv in ds:
        if nv.get('soNG', 0) > 0:
            nv['luongHT'] = 1
        elif nv.get('soSP', 0) > 0:
            nv['luongHT'] = 2
        
        
if __name__ == '__main__':
    ds = []
    khoiTaoNV(ds)
    printNV(ds)
    tinhluongHT(ds)
    printNV(ds)
    
    
```

```python
nhanvien_list = []

def khoiTaoNhanVien(maNV: int, hoTen: str, luongCB: float, soNG=None, soSP=None):
    nhanvien = {
        "maNV": maNV,
        "hoTen": hoTen,
        "luongCB": luongCB
    }
    #nếu tham số soSP được cung cấp, chúng ta gán giá trị của soSP cho khóa "soSanPham" trong từ điển nhanvien
    if soNG is not None:
        nhanvien["soNgayLamViec"] = int(soNG)
    elif soSP is not None:
        nhanvien["soSanPham"] = int(soSP)
    
    if len(nhanvien_list) <= 300:
        nhanvien_list.append(nhanvien)   
        return True
    else:
        #hàm trả về giá trị False để chỉ ra rằng việc thêm nhân viên không thành công.
        return False


def tinhLuong(nhanvien):
    #nếu "soNgayLamViec" có trong nhanvien thì
    if "soNgayLamViec" in nhanvien:
        return nhanvien["luongCB"] + nhanvien["soNgayLamViec"] * 150_000
    elif "soSanPham" in nhanvien:
        return nhanvien["luongCB"] + nhanvien["soSanPham"] * 18_000


def xuatDSNhanVien(ds):
    for nhanvien in ds:
        inThongTin(nhanvien)
        

# def inThongTin(nhanvien):
#     print("Mã nhân viên:", nhanvien["maNV"])
#     print("Họ và tên:", nhanvien["hoTen"])
#     print("Lương cơ bản:", nhanvien["luongCB"])
#     print("Lương hằng tháng:", tinhLuong(nhanvien))
#     print("------------------------")
def inThongTin(nhanvien):
    if "soNgayLamViec" in nhanvien:
        loaiNV = "NVVP"
    elif "soSanPham" in nhanvien:
        loaiNV = "NVBH"
    
    #{:,.0f}: Đây là định dạng cho giá trị số nguyên (f đại diện cho số thực)
    # và ,.0 chỉ định rằng phần ngàn của số sẽ được thêm dấu phân cách hàng nghìn (,) và không có chữ số thập phân (.0)
    luongCB = "{:,.0f} VNĐ".format(nhanvien['luongCB'])
    luongHangThang = "{:,.0f} VNĐ".format(tinhLuong(nhanvien))
    
    #chúng ta sử dụng f-string để tạo chuỗi thongtin với định dạng yêu cầu và in ra thông tin nhân viên theo kiểu hàng ngang.
    thongtin = f"{loaiNV} = [Mã NV: {nhanvien['maNV']}, Họ tên: {nhanvien['hoTen']}, Lương CB: {luongCB}, Lương hằng tháng: {luongHangThang}]"
    print(thongtin)



def timNhanVienTheoMa(maNV):
    for nhanvien in nhanvien_list:
        if nhanvien["maNV"] == int(maNV):
            return inThongTin(nhanvien)
    print("Không tìm thấy nhân viên với mã số {}".format(maNV))


def timNhanVienLuongCaoNhat():
    max_luong = 0
    nhanvien_max_luong = None
    
    for nhanvien in nhanvien_list:
        luong = tinhLuong(nhanvien)
        
        if luong > max_luong:
            max_luong = luong
            nhanvien_max_luong = [nhanvien]
        # nếu nhiều người có lương cao nhất bằng nhau thì in ra hết
        elif luong == max_luong:
            nhanvien_max_luong.append(nhanvien)
    
    print("\n\nDanh sách nhân viên có lương cao nhất:")
    return xuatDSNhanVien(nhanvien_max_luong)


def timNhanVienBanHangLuongThapNhat():
    min_luong = float('inf') #số vô hạn
    nhanvien_min_luong = None
    
    for nhanvien in nhanvien_list:
        if "soSanPham" in nhanvien:
            luong = tinhLuong(nhanvien)
            if luong < min_luong:
                min_luong = luong
                nhanvien_min_luong = [nhanvien]
            # nếu nhiều người có lương thấp nhất bằng nhau thì in ra hết
            elif luong == min_luong:
                nhanvien_min_luong.append(nhanvien)
    
    print("\n\nDanh sách nhân viên bán hàng có lương thấp nhất:")
    return xuatDSNhanVien(nhanvien_min_luong)




if __name__ == '__main__':
    # Khởi tạo các nhân viên
    nhanvien1 = khoiTaoNhanVien(101, "Nguyễn Văn A", 5_000_000, soNG=20)
    nhanvien2 = khoiTaoNhanVien(102, "Trần Thị Thu", 6_000_000, soSP=100)
    nhanvien3 = khoiTaoNhanVien(103, "Lê Hoàng Tuấn", 4_500_000, soNG=25)
    nhanvien4 = khoiTaoNhanVien(104, "Nguyễn Tấn Cầm", 10_000_000, soNG=100)
    nhanvien5 = khoiTaoNhanVien(105, "Lâm Gia Thuận", 6_500_000, soNG=50)
    nhanvien6 = khoiTaoNhanVien(106, "Đinh Tiên Hoàng", 7_500_000, soNG=30)
    nhanvien7 = khoiTaoNhanVien(107, "Trương Minh Quang", 4_200_000, soSP=150)
    nhanvien8 = khoiTaoNhanVien(108, "Cao Đức Hiền", 8_000_000, soSP=130)
    nhanvien9 = khoiTaoNhanVien(109, "Nguyễn Chu Toàn", 7_200_000, soNG=30)
    nhanvien10 = khoiTaoNhanVien(110, "Đinh Thị Mẫn", 7_800_000, soSP=120)
    nhanvien11 = khoiTaoNhanVien(111, "Bùi Thị Hiền", 10_000_000, soNG=100)
    nhanvien12 = khoiTaoNhanVien(112, "Phạm Đức Huy", 10_000_000, soNG=100)
    nhanvien13 = khoiTaoNhanVien(113, "Đào Nhâm Phúc", 4_200_000, soSP=150)

    # Xuất danh sách nhân viên
    xuatDSNhanVien(nhanvien_list)

    
    # Tính lương nhân viên
    maNV = input("\n\nNhập mã số nhân viên cần tính lương: ")
    for nhanvien in nhanvien_list:
        if nhanvien['maNV'] == int(maNV):
            tenNV = nhanvien['hoTen']
            luongNV = tinhLuong(nhanvien)
            print("Nhân viên: {}, Lương: {:,.0f} VNĐ".format(tenNV, luongNV))
            break
    else:
            print("Không tìm thấy nhân viên với mã số {}".format(maNV))

            
    # Tìm nhân viên theo mã
    timNhanVienTheoMa(input("\n\nNhập mã số nhân viên cần tìm: "))

    
    # Tìm nhân viên có lương cao nhất
    timNhanVienLuongCaoNhat()

    # Tìm nhân viên bán hàng có lương thấp nhất
    timNhanVienBanHangLuongThapNhat()
```
