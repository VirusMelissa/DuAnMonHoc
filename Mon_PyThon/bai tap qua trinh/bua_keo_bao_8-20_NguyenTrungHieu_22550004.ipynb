```python
import random as r

def play_game(nguoi_choi):
    #nguoi_choi = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'Pn']
    so_nguoi = len(nguoi_choi)
    sovongchoi = 0; luotchoi = 0
    nguoi_loai = []
    nguoi_thang = []
    nguoi_le = None
    
    # Kiểm tra số người chơi là số lẻ
    if so_nguoi % 2 != 0:
        nguoi_le = nguoi_choi.pop()
    
    # chạy đến khi nào chỉ còn 1 người thắng cuộc
    while len(nguoi_choi) > 1:
        sovongchoi += 1
        print('--------------------------------------')
        print("Vòng:", sovongchoi)
        
        
        # Chọn ngẫu nhiên các cặp chơi
        cap_ngaunhien = []
        # chạy 1 lần 1 cặp, i chạy 0 2 4 6 n thì mỗi lần i đều có người kế tiếp i+1
        # vd nếu có 9 người, thì người lẻ là người cuối cùng i = 8 mà +1 =9 sẽ bằng so_nguoi nên chạy else
        for i in range(0, len(nguoi_choi), 2):
            #kiểm tra xem người chơi kế tiếp có tồn tại hay không
            if i+1 < so_nguoi:
                cap_ngaunhien.append((nguoi_choi[i], nguoi_choi[i+1]))
            else:
                cap_ngaunhien.append((nguoi_choi[i], nguoi_thang[0]))
      
        
        # Chơi từng cặp
        for cap in cap_ngaunhien:
            nguoi1, nguoi2 = cap
            diemso_N1 = 0
            diemso_N2 = 0
            print("\n")
            
            while True:
                luotchoi += 1
                print("\nLượt: {0} của cặp {1} và {2}".format(luotchoi, nguoi1, nguoi2))
                nguoi1_chon = r.choice(['Búa', 'Kéo', 'Bao'])
                nguoi2_chon = r.choice(['Búa', 'Kéo', 'Bao'])
                
                print("Người chơi", nguoi1, "ra:", nguoi1_chon)
                print("Người chơi", nguoi2, "ra:", nguoi2_chon)
                
                if nguoi1_chon == nguoi2_chon:
                    diemso_N1 += 1
                    diemso_N2 += 1
                    print("Hòa")
                #khi đang viết mà xuống dòng thì phải thêm \    
                elif (nguoi1_chon == 'Búa' and nguoi2_chon == 'Kéo') or \
                (nguoi1_chon == 'Kéo' and nguoi2_chon == 'Bao') or \
                (nguoi1_chon == 'Bao' and nguoi2_chon == 'Búa'):
                    diemso_N1 += 1
                    print("Người chơi", nguoi1, "thắng lượt này")
                else:
                    diemso_N2 += 1
                    print("Người chơi", nguoi2, "thắng lượt này")
                
                print("Điểm số: {0} - {1} | {2} - {3}".format(nguoi1, diemso_N1, nguoi2, diemso_N2))
            
                if diemso_N1 == (diemso_N2 + 2):
                    nguoi_loai.append(nguoi2)
                    print("Người chơi", nguoi1, "thắng cặp chơi này ( •̀ ω •́ )✧")
                    luotchoi = 0
                    break
                if diemso_N2 == (diemso_N1 + 2):
                    nguoi_loai.append(nguoi1)
                    print("Người chơi", nguoi2, "thắng cặp chơi này ( •̀ ω •́ )✧")
                    luotchoi = 0
                    break
        
        # Loại bỏ các người chơi đã bị loại
        # tạo một danh sách mới chỉ chứa các phần tử từ danh sách nguoi_choi mà không có trong danh sách nguoi_loai
        nguoi_choi = [nguoichoi_conlai for nguoichoi_conlai in nguoi_choi if nguoichoi_conlai not in nguoi_loai]
        
        # Lưu danh sách người chơi thắng vào vòng tiếp theo
        # Thêm nhiều phần tử vào cuối danh sách khác với append chỉ thêm 1 phần tử
        nguoi_thang.extend(nguoi_choi)
        
        
        nguoi_loai = []
        
        print("\nDanh sách người chơi:")
        for nguoichoi_conlai in nguoi_choi:
            print("Người chơi", nguoichoi_conlai)
        print(nguoi_choi)
        
    print("\nNgười chơi", nguoi_thang[0], "là người chiến thắng cuối cùng!")
    

# Nhập tổng số người chơi từ 8-20
tong_so_nguoi_choi = int(input("Nhập tổng số người chơi (8-20): "))
while tong_so_nguoi_choi < 8 or tong_so_nguoi_choi > 20:
    print("Vui lòng nhập số người chơi từ 8 đến 20.")
    tong_so_nguoi_choi = int(input("Nhập tổng số người chơi (8-20): "))

# Tạo danh sách các người chơi
nguoichoi = []
for i in range(1, tong_so_nguoi_choi + 1): #vì mặc định chỉ chạy từ 0 đến tổng người chơi -1 
    nguoichoi.append("P" + str(i))

# Chơi game
play_game(nguoichoi)
```
